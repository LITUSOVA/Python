```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 5*Y
Z2 = 2*X + 5*Y + 300

plt.title(r' ', fontsize=16, y = 1.05)
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
ax.plot_wireframe(X, Y, Z1,colors='blue', label=' ', linestyle='-', linewidth=1)
ax.plot_wireframe(X, Y, Z2,colors='green',  label=' ', linestyle='-', linewidth=1)
plt.legend(frameon=False)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
