```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
import numpy as np
x = np.linspace(-10, 10, 50)
y_1 = np.sin(x)
y_2 = np.sin(x / 2)
```

```python
plt.plot(x, y_1, color = 'red', marker = 'o', label = 'y = cosx')
plt.plot(x, y_2, color = 'blue', marker = 'x', label = 'y = cos2x')

plt.ylim(-1.25, 2)

plt.grid()

plt.legend(loc = 'upper center', ncol = 2)
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)

plt.title(r'Графики функций $ у = cos(x) $ и $ у = cos(2x)$', fontsize=16, y = 1.05)
```
