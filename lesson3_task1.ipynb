```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 300)
k1, a1, b1 = np.linspace(1, 10, num = 3)
k2, a2, b2 = np.linspace(11, 20, num = 3)
k3, a3, b3 = np.linspace(21, 30, num = 3)


plt.figure(figsize = (20, 10))
plt.grid()
plt.ylim(0, 60)

plt.title(r'Графики $ y(x) = k*cos(x – a) + b $', fontsize=16, y = 1.05)
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
    
plt.plot(x, k1 * np.cos(x - a1) + b1, color='b', label = 'Для k1, a1, b1')
plt.plot(x, k2 * np.cos(x - a2) + b2, color='g', label = 'Для k2, a2, b2')
plt.plot(x, k3 * np.cos(x - a3) + b3, color='r', label = 'Для k3, a3, b3')
plt.legend(loc = 'upper center', ncol = 14, fontsize=16)

plt.show()
```
